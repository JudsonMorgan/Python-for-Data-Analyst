```python
#reordering of the columns
data.columns.values
```

```python
columns = ['State', 'Month', 'Year', 'Dayweek', 'Time',
       'Bus Involvement', 'Heavy Rigid Truck Involvement',
       'Articulated Truck Involvement', 'Speed Limit', 'Road User',
       'Gender', 'Age', 'Christmas Period', 'Easter Period', 'Age Group',
       'Day of week', 'Time of day', 'Crash Type']

data = data[columns]
data.head()
```

### Target Encodding

```python
data['Crash Type'] = data['Crash Type'].apply(lambda x: 0 if x == "Single" else 1)

data['Crash Type'].value_counts()
```

```python
data.head()
```

```python
sns.scatterplot(data = data, x = 'Age', y='Crash Type')
```
